## Perform a statistical test

### Average response time across complaint types is similar or not (overall)
From the above data, its observed that the average resolution time across the complaint types are not equal.

But, the following complaint types have response time which were too close.

Traffic 12304.383994 Disorderly Youth 12363.749206

Null Hypothesis(H0) - Average response time across complaint types are equal Alternate Hypothesis(H1) - Average response time across complaint types are not equal

```python
df_traffic = df[df['Complaint Type']=='Traffic']
```

```python
df_traffic = df_traffic.loc[ : ,'Request_Closing_Time']
```

```python
df_traffic.head()
```

```python
df_dis_youth = df[df['Complaint Type']=='Disorderly Youth']
```

```python
df_dis_youth = df_dis_youth.loc[ : ,'Request_Closing_Time']
```

```python
df_dis_youth.head()
```

```python
fvalue, pvalue = stats.f_oneway(df_traffic, df_dis_youth)
```

```python
pvalue
```

  Accept null hypothesis(H0) for complaint types 'Traffic' and 'Disorderly Youth',as the pvalue > 0.05

```python
fvalue
```

### Using crosstab and Chi square test to check if the complaints and location are related

```python
city_compaint = pd.crosstab(df['City'], df['Complaint Type'])
```

```python
city_compaint.head()
```

```python

from scipy.stats import chi2_contingency
```

```python
#contingency table
table = city_compaint
```

# Get chi-square value , p-value, degrees of freedom, expected frequencies using the function chi2_contingency

```python
chi_square_value, p_value, degree_of_freedom, expected_frequency = chi2_contingency(table)
```

```python
chi_square_value
```

```python
p_value
```

```python
degree_of_freedom
```

```python
 expected_frequency
```

```python
# select significance value
alpha = 0.05

# Determine whether to reject or keep your null hypothesis
print('significance=%.3f, p=%.3f' % (alpha, p_value))
if p_value <= alpha:
    print('Variables are associated (reject H0)')
else:
    print('Variables are not associated(fail to reject H0)')
```
